```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot as plt
import os
import cv2

loc = r"C:\Users\bolt6281\Desktop\smartcycle\SmartCycle_AI\SmartCycle_Trainer\data augmentation\augmented_data"

for each in data_list:

    if each[-1] == "1":
        brightness_lowest = 0.6
        to_be = 1200
    else:
        brightness_lowest = 0.8
        to_be = 800
        
    # create image data augmentation generator
    datagen = ImageDataGenerator(horizontal_flip=True, rotation_range=35, brightness_range=[brightness_lowest,1.0])
        
    inside = loc + "\\" + each
    inside_list = os.listdir(inside)[:-1]
    
    while len(os.listdir(inside)[:-1]) < to_be:
        
        for i, inside_each in enumerate(inside_list):
            
            # load the image
            img = load_img(inside + "\\" + inside_each)

            # convert to numpy array
            data = img_to_array(img)

            # expand dimension to one sample
            samples = expand_dims(data, 0)

            # prepare iterator
            it = datagen.flow(samples, batch_size=1)

            # generate samples
            batch = it.next()[0].astype('uint8')
            cv2.imwrite(inside + "\\" + inside_each + "_argumented_" + i + ".jpg", batch)
```
